```python
import zipfile
import os

def unzip_folder(zip_file_path, extract_to_folder):
    # Create the folder if it doesn't exist
    os.makedirs(extract_to_folder, exist_ok=True)

    # Open the zip file
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        # Extract all the contents into the specified folder
        zip_ref.extractall(extract_to_folder)
        print(f'Extracted all files to {extract_to_folder}')

# Example usage
zip_file = '/content/PlateDetectionYolov8.v9i.yolov9.zip'
extract_folder = '/content/PlateDetectionYolov8.v9i.yolov9'
unzip_folder(zip_file, extract_folder)
```

```python
import os
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import random

# Path to test images and labels folder
images_folder = "/content/PlateDetectionYolov8.v9i.yolov9/test/images"
labels_folder = "/content/PlateDetectionYolov8.v9i.yolov9/test/labels"

color_map = {}  # Map for storing unique colors per class

# Helper function to generate random colors
def random_color():
    return (random.random(), random.random(), random.random())

# Function to display images with bounding boxes for all classes
def visualize_images_all_classes(num_images=6):
    images = os.listdir(images_folder)
    random.shuffle(images)
    selected_images = images[:num_images]

    fig, axes = plt.subplots(1, len(selected_images), figsize=(20, 5))
    for i, img_file in enumerate(selected_images):
        img_path = os.path.join(images_folder, img_file)
        label_path = os.path.join(labels_folder, img_file.replace(".jpg", ".txt"))

        # Load and display image
        image = Image.open(img_path)
        axes[i].imshow(image)
        axes[i].axis('off')

        # Draw bounding boxes from label file
        if os.path.exists(label_path):
            with open(label_path, 'r') as f:
                for line in f:
                    data = line.strip().split()
                    class_id = int(data[0])

                    # Assign a color for the class if not already assigned
                    if class_id not in color_map:
                        color_map[class_id] = random_color()

                    # Parse bounding box coordinates
                    x_center, y_center, width, height = map(float, data[1:])
                    img_width, img_height = image.size
                    x_min = (x_center - width / 2) * img_width
                    y_min = (y_center - height / 2) * img_height
                    box_width = width * img_width
                    box_height = height * img_height

                    # Draw rectangle on the axes
                    rect = patches.Rectangle((x_min, y_min), box_width, box_height,
                                             linewidth=2, edgecolor=color_map[class_id], facecolor='none')
                    axes[i].add_patch(rect)
                    axes[i].text(x_min, y_min, str(class_id), color='white', fontsize=10,
                                 bbox=dict(facecolor=color_map[class_id], edgecolor='none', alpha=0.7))
    plt.tight_layout()
    plt.show()

# Call the function to visualize images with all classes
visualize_images_all_classes(num_images=6)
```

```python
import os
import random
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
from ultralytics import YOLO

# Helper function to generate random colors for each class
def random_color():
    return (random.random(), random.random(), random.random())

def infer_and_visualize(image_path, model_path='yolov9s.pt'):
    """
    Loads a YOLO model, runs inference on an image, and visualizes the results.

    Parameters:
        image_path (str): Path to the input image.
        model_path (str): Path to the YOLO model weights (default is 'yolov9s.pt').
    """
    # Load the YOLO model
    model = YOLO(model_path)

    # Run inference on the image.
    # The model call returns a list of result objects (one per image).
    results = model(image_path)
    result = results[0]  # Since we're processing one image

    # Open the image using PIL and convert it to a NumPy array for plotting.
    image = Image.open(image_path)
    image_np = np.array(image)

    # Set up the Matplotlib figure and axis.
    fig, ax = plt.subplots(1, figsize=(10, 10))
    ax.imshow(image_np)

    # Create a dictionary to store a unique color per class.
    color_map = {}

    # Check if detections exist in the result.
    if hasattr(result, 'boxes') and result.boxes is not None:
        # Retrieve bounding boxes, confidence scores, and class indices.
        # The boxes are in [x1, y1, x2, y2] format.
        # Note: Depending on your YOLO version, you may need to adjust these attributes.
        boxes = result.boxes.xyxy.cpu().numpy() if hasattr(result.boxes.xyxy, "cpu") else result.boxes.xyxy.numpy()
        confs = result.boxes.conf.cpu().numpy() if hasattr(result.boxes.conf, "cpu") else result.boxes.conf.numpy()
        classes = result.boxes.cls.cpu().numpy() if hasattr(result.boxes.cls, "cpu") else result.boxes.cls.numpy()

        # Loop over each detection and draw the bounding box and label.
        for i, box in enumerate(boxes):
            x1, y1, x2, y2 = box
            class_id = int(classes[i])
            conf = confs[i]

            # Get or assign a color for this class.
            if class_id not in color_map:
                color_map[class_id] = random_color()

            # Create a rectangle patch for the bounding box.
            rect = patches.Rectangle((x1, y1), x2 - x1, y2 - y1,
                                     linewidth=2,
                                     edgecolor=color_map[class_id],
                                     facecolor='none')
            ax.add_patch(rect)

            # Create a label string with the class id and confidence score.
            label = f"{class_id}: {conf:.2f}"
            ax.text(x1, y1, label,
                    fontsize=12,
                    color='white',
                    bbox=dict(facecolor=color_map[class_id], edgecolor='none', alpha=0.5))
    else:
        print("No detections found.")

    plt.axis('off')
    plt.tight_layout()
    plt.show()

# Example usage:
# Replace the image path with the path to your test image.
image_path = "/content/Plate-Sample.jpg"
# If you are training the model from the beginning, use this path.
# infer_and_visualize(image_path, model_path='/content/plate_detection/yolov9_plate_model/weights/best.pt')
# If you just load the best weights, use this path.
infer_and_visualize(image_path, model_path='/content/best.pt')
```
