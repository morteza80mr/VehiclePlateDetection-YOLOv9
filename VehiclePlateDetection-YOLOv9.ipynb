```python
import zipfile
import os

def unzip_folder(zip_file_path, extract_to_folder):
    # Create the folder if it doesn't exist
    os.makedirs(extract_to_folder, exist_ok=True)

    # Open the zip file
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        # Extract all the contents into the specified folder
        zip_ref.extractall(extract_to_folder)
        print(f'Extracted all files to {extract_to_folder}')

# Example usage
zip_file = '/content/PlateDetectionYolov8.v9i.yolov9.zip'
extract_folder = '/content/PlateDetectionYolov8.v9i.yolov9'
unzip_folder(zip_file, extract_folder)
```

```python
import torch
print(torch.cuda.is_available())  # Should return True if CUDA is properly installed
```

```python
from ultralytics import YOLO

# Load the YOLO model
model = YOLO('yolov9s.pt')  # Start with a pretrained YOLOv9 model

# Train the model
results = model.train(
    data='/content/PlateDetectionYolov8.v9i.yolov9/data.yaml',         # Path to data.yaml
    epochs=50,                # Number of epochs
    batch=16,                 # Batch size
    imgsz=224,                # Image size
    project='plate_detection', # Save results to this folder
    name='yolov9_plate_model', # Name of the experiment
    device=0              # Use CPU for training
)

# Print the results
print("Training completed. Check the 'plate_detection' folder for results.")
```

```python
results = model.val(data='/content/PlateDetectionYolov8.v9i.yolov9/data.yaml')
print(results)
```
